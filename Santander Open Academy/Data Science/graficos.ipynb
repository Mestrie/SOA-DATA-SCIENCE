```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Best_Movies_Netflix.csv')
df = df.drop(df.columns[0], axis= 1)
```

```python

df.head()
```

```python
eixo_x = df['MAIN_GENRE'].value_counts().keys()
eixo_y = df['MAIN_GENRE'].value_counts().values
```

```python
plt.figure(figsize=(20, 12))

plt.bar(eixo_x, eixo_y)

plt.title('Filmes por Genero')

plt.xticks(rotation=45)

plt.xlabel('Generos')
plt.ylabel('Quantidade')

plt.show()
```

```python
eixo_x2 = df['RELEASE_YEAR'].value_counts().sort_index().keys()
eixo_y2 = df['RELEASE_YEAR'].value_counts().sort_index().values
```

```python
plt.figure(figsize=(18, 10))
plt.plot(eixo_x2, eixo_y2)

plt.title('Melhores Filmes por Ano de Lanamento')
plt.xlabel('Ano de Lanamento')
plt.ylabel('Quantidade')
plt.show()
```
